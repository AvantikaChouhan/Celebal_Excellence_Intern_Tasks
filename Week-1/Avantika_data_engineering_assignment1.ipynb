```python
import zipfile
with zipfile.ZipFile("shopping_dataset.zip",'r') as zip_ref:
    zip_ref.extractall("data")
```

```python
import pandas as pd
import os
all_files=os.listdir("data")
dfs=[]
for file in all_files:
    if file.endswith(".csv"):
        df=pd.read_csv("data/"+file)
        df["category"]=file
        dfs.append(df)
final_df=pd.concat(dfs)
```
